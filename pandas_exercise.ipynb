Exercise 10: Display the movies longer than 30 minutes and shorter than 360 minutes

```python
data_10 = pd.read_csv("https://raw.githubusercontent.com/Laxminarayen/Inceptz-Batch13-Analytics_and_Python/master/05%20-%20Day%205%20-%20Pandas%20with%20Exercises/IMDB-Movie-Data.csv")
data_10
```

```python
print("List of movies longer than 30 minutes and shorter than 360 minutes:")
new_movie = data_10.loc[:,['Title','Runtime (Minutes)']]

min_10 = new_movie.loc[(new_movie['Runtime (Minutes)'] > 30) & (new_movie['Runtime (Minutes)'] < 360)]
                      
```
